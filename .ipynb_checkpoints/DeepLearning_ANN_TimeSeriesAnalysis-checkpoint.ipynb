<b>Plotting features to identify Missing Data and Outliers Manually</b></br>
<p>Of course you can use your own methods as ECDF, Statistical Analysts, Etc.</p>

```python
%%time

# Simple function to plot any feature from dataset in a given Timeline
def plotTimeSeries(dataset, feature_name):
    _ = plt.rcParams['figure.figsize'] = (15, 5)
    _ = plt.grid(which = 'major', c = '#cccccc', alpha = 0.5)
    _ = plt.scatter(dataset.index, dataset[feature_name],
                    marker = 'o', s = 0.1, alpha = 0.5,
                    color = 'blue')
    _ = plt.title(feature_name, family = 'Arial', fontsize = 16)
    _ = plt.xlabel('Timestamp [Date Time]', family = 'Arial', fontsize = 14)
    _ = plt.ylabel(feature_name, family = 'Arial', fontsize = 14)
    _ = plt.xticks(fontsize = 14)
    _ = plt.yticks(fontsize = 14)
    _ = plt.show()
                  
    return None

# Simple way to remove data outliers
def removeOutliers(dataset, feature_name):
    data_temp = dataset[feature_name]
    
    d_min = np.min(data_temp)
    d_mean = np.mean(data_temp)
    
    data_temp = data_temp.replace(d_min, d_mean)
    dataset[feature_name] = data_temp
    
    return dataset

# Remove data outliers (if necessary, else comment following lines)
dataset = removeOutliers(dataset, 'wv (m/s)')
dataset = removeOutliers(dataset, 'max. wv (m/s)')
    
# Plot all features in Timeline
for feature in columns:
    plotTimeSeries(dataset, feature)
    
```

<h3>Data Generator</h3>
<p>The data-set now been prepared as 2-dimensional Numpy array. The training-data has appr. 370k observations, consisting of 14 input-signals and 4 output-signals.</p></br>
<p>These are the array-shaped input and output data:</p>

```python
print(x_train_scaled.shape)
print(y_train_scaled.shape)
```

<p>Instead of training the Recurrent Neural Network on the complete sequences of appr. 370k observations we will use the following function to create a batch of shorter sub-sequences picked at random from the training-data.</p>

```python
def batch_generator(batch_size, sequence_lenght):
    """
    Generator function for creating a Random Batches of Training-data
    """
    
    # Infinite loop (!)
    while True:
        # Allocate a new array for the batch of Input-signals.
        x_shape = (batch_size, sequence_lenght, num_x_signals)
        x_batch = np.zeros(shape = x_shape, dtype = np.float16)
        
        # Allocate a new array for the batch of Output-signals.
        y_shape = (batch_size, sequence_lenght, num_y_signals)
        y_batch = np.zeros(shape = y_shape, dtype = np.float16)
        
        # Fill the batch with random sequences of data.
        for i in range(batch_size):
            # Get a randon start-index
            # This points somewhere into the training-data
            idx = np.random.randint(num_train - sequence_lenght)
            
            # Copy the sequence of data starting at this index
            x_batch[i] = x_train_scaled[idx : idx + sequence_lenght]
            y_batch[i] = y_train_scaled[idx : idx + sequence_lenght]
            
        yield(x_batch, y_batch)
```

<p>I will use a quite large batch-size so as to keep the GPU near 100% workload. You may have to adjust this number depending on your GPU, its RAM and your choice of <code>sequence_lenght</code> below.</p>

```python
batch_size = 64
```

<p>We will use a sequence-lenght of <b>12 960</b>, which means that each random sequence contains observations for <b>8</b> weeks. One time-step correspond to 10 minutes, so <code>6×24</code> correspond to 1 days and <code>6×24×10</code> corresponds to 10 days.</p>

```python
sequence_lenght = 6 * 24 * 30 * 3
print('Sequence lenght is {} observations.'.format(sequence_lenght))
```

<p>Creating the Batch-Generator.</p>

```python
generator = batch_generator(batch_size = batch_size,
                            sequence_lenght = sequence_lenght)
```

```python
x_batch, y_batch = next(generator)
```

<h2>Create the Recurrent Neural Network</h2>
<p>We are now ready to crate the Recurrent Neural Network (RNN). We will use the <i>Keras API</i> for this because of it's simplicity.</p>

```python
model = Sequential()
```

<p>We can now add Gated Recurrent Unit (GRU) to the network. This will have 512 outputs for each time-step in the sequence.</p></b>
<p>Note that because this is the first layer in the model, Keras needs to know the shape of its input, which is a batch of sequences of arbitrary lenght (indicated by <code>None</code>), where each observation has a number of input-signals (<code>num_x_signals</code>).</p>

```python
model.add(GRU(units = 64, # modified: 512
              return_sequences = True,
              input_shape = (None, num_x_signals,)))
```

<p>The <i>GRU</i> outputs a batch of sequences of <b>512</b> values. I want to predict <b><u>4</u></b> output-signals, so I add fully-connected (or dense) layer which maps <b>512</b> values down to only <b>4</b> values.</p></br>
<p>The output-signals in the data-set have been limited to be between <code>0</code> and <code>1</code> using a scaler-object. So I also limit the output of the Neural Network using the <i>Sigmoid</i> activation function, which squashes the output to be between <code>0</code> and <code>1</code>.</p>

```python
model.add(Dense(num_y_signals, activation = 'sigmoid'))
```

<img src="img/sigmoid.png">

<p>A problem with using the Sigmoid activation function, is that I can now only output values in the same range as the training-data.</p></br>
<p>For example, if the training-data only has temperature between <code>-20</code> and <code>+30</code> degrees, then scaler-object will map <code>-20</code> to <code>0</code> and <code>+30</code> to <code>1</code>. So if I limit the output of the neural network to be between <code>0</code> and <code>1</code> using the Sigmoid function, this can only be mapped back to temperature values between <code>-20</code> and <code>+30</code>.</p></br>
<p>We can use a linear activation function on the output instead. This allows for the output to take on arbitrary values. It might work with the standard initalization for a simple network architecture, but for more complicated network architectures e.g. with more layers, it might be necessary to initialize the weights with smaller values to avoid <code>NaN</code> values during training. You may need to experiment with this to get it working.</p>

```python
if False:
    from tensorflow.python.keras.initializers import RandomUniform
    
    # Maybe use lower init-ranges
    init = RandomUniform(minval = -0.05, maxval = 0.05)
    
    model.add(Dense(num_y_signals,
                    activation = 'linear',
                    kernel_initializer = init))
```

<h3>Loss Function</h3></br>
<p>I will use Mean Squared Error (<b>MSE</b>) as the loss-function that will be minimized. This measures how closely the model's output matches the true output signals.</p></br>
<img src="img/mse.jpg"></br>
<p>Hovewer, at the begining of a sequence, the model has only seen input-signals for a few time-steps, so its generated output may be very inaccurate. Using the loss-value for the early time-steps may cause the model to distort its later output. I therefore give the model a <i>warmup-period</i> of 50 time-step where I don't use its accuracy in the loss-function, in hope of improving the accuracy for later time-steps.</p>

```python
warmup_steps = 50
```

```python
def loss_mse_warmup(y_true, y_pred):
    """
    Calculate the Mean Squared Error between y_true and y_pred,
    but ignore the begining "warmup" part of sequences.
    
    y_true : desired output.
    y_pred : model's output.
    """
    
    # The shape of both input tensors are:
    # [batch_size, sequence_lenght, num_y_signals].
    
    # Ignore the "warmup" parts of the sequences
    # by taking slices of the tensors.
    y_true_slice = y_true[:, warmup_steps:, :]
    y_pred_slice = y_pred[:, warmup_steps:, :]
    
    # These sliced tensors both have this shape:
    # [batch_size, sequence_lenght - warmup_steps, num_y_signals]
    
    # Calculate the MSE loss for each value in these tensors.
    # This outputs a 4-rank tensor of the same shape.
    loss = tf.losses.mean_squared_error(labels = y_true_slice,
                                        predictions = y_pred_slice)
    
    # Keras may reduce this across the first axis (the batch)
    # but the semantics are unclear, so to be sure we use
    # the loss across the entire tensor, I reduce it to a 
    # single scalar with the mean function.
    loss_mean = tf.reduce_mean(loss)
    
    return loss_mean
```

<h3>Compile Model</h3>
<p>This is the optimizer and the begining learning-rate that I will use.</p>

```python
optimizer = RMSprop(lr = 1e-3)
```

<p>I then compile the Keras model so it is ready for training.</p>

```python
model.compile(loss = loss_mse_warmup, optimizer = optimizer)
```

<p>This is very small model only with two layers. The output of (<code>None</code>, <code>None</code>, <code>3</code>) means that the model will output a batch with an arbitrary number of sequences, each of which has an arbitrary number of observation, and each obsevation has <code>4</code> signals. This correspond to the <code>4</code> target signals we want to predict.</p>

<h3>Callback Functions</h3>
<p>During traiing I want to save checkpoints and log the progress to TensorBoard so we create the appropriate callbacks for Keras. This is the callback for writing checkpoints during training.</p>

```python
path_checkpoint = '23_checkpoint.keras'
callback_checkpoint = ModelCheckpoint(filepath = path_checkpoint,
                                      monitor = 'val_loss',
                                      verbose = 1,
                                      save_weights_only = True,
                                      save_best_only = True)
```

<p>This is the callback for stopping the optimization when performance worsens on the <code>validation-set</code>.</p>

```python
callback_early_stopping = EarlyStopping(monitor = 'val_loss',
                                        patience = 5,
                                        verbose = 1)
```

<p>This is the callback for writing <code>TensoBoard</code> log during training.</p>

```python
callback_tensorboard = TensorBoard(log_dir = './23_logs/',
                                   histogram_freq = 0,
                                   write_graph = False)
```

<b>ISSUE of <i>AttributeError: module 'pandas.core.computation' has no attribute 'expressions</i></b> - https://github.com/pandas-dev/pandas/issues/18355

<p>This callback reduces the learning-rate for the optimizer of the validation-loss has not improved since the last epoch (as indicated by <code>patience = 0</code>. The learning-rate will be reduced by multiplying it with given factor. We set a start learning-rate of <code>1e-3</code> above, so multiplying it by <code>0.1</code> gives a learning rate of <code>1e-4</code>. I don't want the learning-rate to go any lower than this.</p>

```python
callback_reduce_lr = ReduceLROnPlateau(monitor = 'val_loss',
                                       factor = 0.1,
                                       min_lr = 1e-4,
                                       patience = 0,
                                       verbose = 1)
```

```python
callbacks = [callback_early_stopping,
             callback_checkpoint,
             callback_tensorboard,
             callback_reduce_lr]
```

<h2>General Predictions</h2>
<p>This helper-function plot the predicted and true output-signals.</p>

```python
def plot_comparison(start_idx, lenght = 100, train = True):
    
    """
    Plot the predicted and true output-signals.
    
    Parameters:
        > start_idx: Start-index for the time-series
        > lenght: Sequence-lenght to process and plot
        > train: Boolean whether to use training or test-test.
    """
    
    if train:
        # Use training-data
        x = x_train_scaled
        y_true = y_train
    
    else:
        # Use test-data
        x = x_test_scaled
        y_true = y_test
        
    # End-index for the sequences.
    end_idx = start_idx + lenght
    
    # Select the sequences from the given start-index and of the
    # given lenght.
    x = x[start_idx:end_idx]
    y_true = y_true[start_idx:end_idx]
    
    # Input-signals for the model.
    x = np.expand_dims(x, axis = 0)
    
    # Use the model to predict the output-signals.
    y_pred = model.predict(x)
    
    # The output of the model is between 0 and 1.
    # Do an inverse map tp get it back to the scale
    # of the original data-set (!!!)
    y_pred_rescaled = y_scaler.inverse_transform(y_pred[0])
    
    # For each output-signal.
    for signal in range(len(target_features)):
        
        # Get the output-signal predicted by the model.
        signal_pred = y_pred_rescaled[:, signal]
        
        # Get the true output-signal from the data-set
        signal_true = y_true[:, signal]
        
        # Make the plotting-canvas bigger.
        plt.figure(figsize = (15, 5))
        
        # Plot and compare the two signals.
        plt.plot(signal_true, label = 'True', color = 'green')
        plt.plot(signal_pred, label = 'Pred.', color = 'blue')
        
        # Plot grey box for warmup-period.
        p = plt.axvspan(0, warmup_steps, facecolor = 'black', alpha = 0.15)
        
        # Plot labels and other components.
        plt.ylabel(target_features[signal], fontsize = 14)
        plt.legend(loc = 'best', shadow = True)
        plt.show()
        
    return None
```

<p>We can now plot an example of predicted output-signals. It is important to understand what these plots show, as they are actually a bit more complicated than you might think.</p></br>

<p>These plots only show the output-signals and not the 14 input-signals used to predict the output-signals. The time-shift between the input-signals and the output-signals is held fixed in these plots. The model <b>always</b> predicts the output-signals e.g. 24 hours into the future (as defined in the shift_steps variable above). So the plot's x-axis merely shows how many time-steps of the input-signals have been seen by the predictive model so far.</p></br>

<p>The prediction is not very accurate for the first 30-50 time-steps because the model has seen very little input-data at this point. The model generates a single time-step of output data for each time-step of the input-data, so when the model has only run for a few time-steps, it knows very little of the history of the input-signals and cannot make an accurate prediction. The model needs to "warm up" by processing perhaps 30-50 time-steps before its predicted output-signals can be used.</p></br>

<p>That is why we ignore this "warmup-period" of 50 time-steps when calculating the mean-squared-error in the loss-function. The "warmup-period" is shown as a grey box in these plots.</p></br>

<p>Let us start with an example from the training-data. This is data that the model has seen during training so it should perform reasonably well on this data.</p>

```python
plot_comparison(start_idx = 144 * 10, lenght = 144 * 14, train = True)
```

<p>The model was able to predict the overall oscillations of the temperature quite well but the peaks were sometimes inaccurate. For the wind-speed, the overall oscillations are predicted reasonably well but the peaks are quite inaccurate. For the atmospheric pressure, the overall curve-shape has been predicted although there seems to be a slight lag and the predicted curve has a lot of noise compared to the smoothness of the original signal.</p>
